```python
from wordcloud import WordCloud

from matplotlib import pyplot as plt

path = 'F:\CodeHub\PythonWorkSpace\ipynb\lib\.txt'

font = r'C:\Windows\Fonts\simfang.ttf'''

f = open(path,encoding='UTF-8')

txt = f.read()


txt=txt.replace('','')

txt=txt.replace('','')


# (maskstopwordsmax_font_size)

wd = WordCloud(background_color = 'black', width = 1000, height = 500,margin = 0,font_path=font, max_font_size=100)


# 

wd_cut = wd.generate(txt)


# plt

plt.imshow(wd_cut,interpolation="bilinear")


# 

plt.axis("off")

# plt.figure(figsize=(100,100),dpi=100)


plt.savefig('123.jpg')

# wd_cut.to_file('wd.jpg')

# 


plt.show()
```
