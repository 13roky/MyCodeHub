```python
from wordcloud import WordCloud

from matplotlib import pyplot as plt

path = 'F:\CodeHub\PythonWorkSpace\ipynb\lib\三国演义.txt'

font = r'C:\Windows\Fonts\simfang.ttf'''

f = open(path,encoding='UTF-8')

txt = f.read()


txt=txt.replace('刘皇叔','刘备')

txt=txt.replace('翼德','张飞')


# 创建词云对象(增加mask背景参数，stopwords停用词，max_font_size最大字体)

wd = WordCloud(background_color = 'black', width = 1000, height = 500,margin = 0,font_path=font, max_font_size=100)


# 使用词云对文本进行分词和词云的生成

wd_cut = wd.generate(txt)


# plt生成词云图片

plt.imshow(wd_cut,interpolation="bilinear")


# 关闭坐标轴

plt.axis("off")

# plt.figure(figsize=(100,100),dpi=100)


plt.savefig('123.jpg')

# wd_cut.to_file('wd.jpg')

# 交互式展示词云图


plt.show()
```
